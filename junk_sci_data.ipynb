```python
import re  # regular expressions module: will use to remove extra stuff from texts
import nltk #natural language toolkit for NLP tasks
# collection of datasets and resources, like lists of words or texts (corpus is module in nltk)
from nltk.corpus import stopwords # list of common words to filter out
```

```python
nltk.download('stopwords') #download
stop_words = set (stopwords.words("english")) #creating a set of english stopwords
```

```python
# extracting the  "content" field from rows labeled as junk science, dropping missing texts as well
junk_science_texts = junksci_df[junksci_df["type"] == "junksci"]["content"].dropna().tolist()
```

```python
def preprocess_text(text):
    text = text.lower()  #convert to lowercase 
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)  #remove URLs
    text = re.sub(r'[^a-z\s]', '', text)  #remove all non-letter characters (punctuation, digits)

    words = text.split() #spliting text into a list of words
    #filter out the stop words
    tokens = []
    for word in text.split():
        if word not in stop_words:
            tokens.append(word)

    return " ".join(tokens)  # join cleaned words back together
```

```python
junk_science_texts
```

```python
# Text preprocessing on all junk science articles
cleaned_junk_science_articles = []

for article in junk_science_texts:
    cleaned_article = preprocess_text(article)
    cleaned_junk_science_articles.append(cleaned_article)

cleaned_junk_science_articles
```

***TOPIC MODELING USING FULL DATASET***

```python
import pandas as pd

dfs = [] #list to store the dataframes
chunk_size = 10000
total_rows = 0
N = 1000000  #rows in datafram

for chunk in pd.read_csv(
    "news_cleaned_2018_02_13.csv",
    chunksize=chunk_size,
    engine="python",
    on_bad_lines="skip"
):
    dfs.append(chunk)
    total_rows += len(chunk)
    if total_rows >= N:
        break

df = pd.concat(dfs, ignore_index=True)  #final with all parts
print(df.head())
```
